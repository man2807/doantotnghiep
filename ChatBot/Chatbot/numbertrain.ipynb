```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain = []
ytrain = []
number = [0] * 100
date = [0]*10
# arr = np.array([[1, 2, 3, 4,8,6,7],[4, 3, 3, 3, 5,6,7],[5, 3, 3, 4, 5,6,7],[9, 3, 3, 4, 5,6,7],[6, 3, 3, 4, 5,6,7],[1, 3, 3, 4, 5,6,7],[1, 3, 3, 4, 5,6,7]])
# y = np.array([1, 2, 3, 4, 5,6,7])
data = pd.read_csv('C:/Users/Admin/Desktop/number/numbersau.csv')
X = list(data.values[:,2:])
y = list(data.values[:,1])
for j in X:
    j = list(j)
    for idxnum,num in enumerate(number):        
        for l in j:
            if(l == idxnum):
                number[idxnum] = 1
    xtrain.append(number)
    number = [0] * 100
```

```python
for jy in y:
    for idxy,numy in enumerate(date):
        if(jy == idxy):
            date[idxy] = 1
            ytrain.append(date)
            date = [0]*10
```

```python

from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.optimizers import gradient_descent_v2
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
xtrain = list(xtrain)
ytrain = list(ytrain)

reduce_lr = ReduceLROnPlateau(monitor="loss", factor=0.1, patience=10, verbose=1)

early_stopping = EarlyStopping(
    monitor="loss", min_delta=0, patience=20, verbose=1
)

model = Sequential()
model.add(Dense(128, input_shape=(len(xtrain[0]),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(ytrain[0]), activation='softmax'))

# Compile model. Stochastic gradient descent with Nesterov accelerated gradient gives good results for this model
sgd = gradient_descent_v2.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
model.summary()
#fitting and saving the model 
print(xtrain[3])
xtrain = np.asarray(xtrain).astype(np.float32)
ytrain = np.asarray(ytrain).astype(np.float32)
hist = model.fit(xtrain, ytrain, epochs=200, batch_size=5, verbose=1,callbacks=[reduce_lr, early_stopping])
model.save('C:/Users/Admin/Desktop/number/numbersau____.h5', hist)

print("model created")
```
